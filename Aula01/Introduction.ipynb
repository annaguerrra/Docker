## Docker
É um sistema de gerenciamento de containers, que depende um kernel Linux, cada container roda uma única aplicação com um ambiente mínimo para a aplicação, sem Ambiente. É o conjunto de dependências necessárias para manter a aplicação:
- Bibliotecas
- Runtime (ex: .Net, Node, Java)
- Variáveis de ambiente
- Estrutura de arquivos necessária

Linux não é um sistema operacional utilizável por usuários ocmunds, é um kernel.

## Container
É um processo isolado, definindo um novo diretório root (/) para ele, utilizando namespaces

# Docker Compose
é uma ferramenta do Docker que permite definir, iniciar e gerenciar multiplos contêineres ao mesmo tempo como um único serviço.

Em vez de executar vários comandos manuais para configurar cada contêiner, rede ou volume, você utiliza um único arquivo de configuração para descrever toda a sua infraestrutura. Com um único comando, todo o ambiente da sua aplicação é criado e interligado automaticamente.
